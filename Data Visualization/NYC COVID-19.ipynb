```python

X = df2['AGE_GROUP']
Y = df2['COVID_CASE_RATE']
Z = df2['HOSPITALIZED_CASE_RATE']
W = df2['DEATH_RATE']


def subcategorybar(X, vals, width=0.8):
    n = len(vals)
    _X = np.arange(len(X))
    for i in range(n):
        plt.bar(_X - width/2. + i/float(n)*width, vals[i], 
                width=width/float(n), align="edge")   
    
    
    plt.xticks(rotation=45, ha='right')
    plt.xticks(_X, X)

subcategorybar(X, [Y,Z,W])
plt.legend(['COVID_CASE_RATE', 'HOSPITALIZED_CASE_RATE', 'DEATH_RATE'],bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```
